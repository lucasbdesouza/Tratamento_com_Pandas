## Criando novas Variáveis

Foi solicitado para adicionar 3 novas variáveis, o Valor Bruto, o Valor por metro quadrado e o Valor Bruto por metro quadrado.

```python
dados_limpos
```

```python
dados_limpos['Valor Bruto'] = dados_limpos['Valor'] + dados_limpos['Condominio'] + dados_limpos['IPTU']
```

```python
dados_limpos
```

```python
dados_limpos['Valor m2'] = dados_limpos['Valor'] / dados_limpos['Area']
dados_limpos
```

```python
dados_limpos['Valor m2'] = dados_limpos['Valor m2'].round(2)
dados_limpos
```

```python
dados_limpos['Valor Bruto m2'] = (dados_limpos['Valor Bruto'] / dados_limpos['Area']).round(2)
dados_limpos
```

Agora foi solicitado para criar uma coluna e classificar se o imóvel é Casa ou Apartamento

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados_limpos['Tipo Agregado'] = dados_limpos['Tipo'].apply(lambda x:'Casa' if x in casa else 'Apartamento')
```

## Análises Descritivas

Foi solicitado pelo Gestor para criar um DataFrame apenas com as informações da média de Valor, da Área e do Valor por metro quadrado. Apenas dos seguintes bairros: 'Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca'

```python
dados_limpos.describe()
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
```

```python
grupo_bairro = dados_limpos['Bairro'].isin(bairros)
novos_dados = dados_limpos[grupo_bairro]
novos_dados
```

```python
novos_dados_agrupados = novos_dados.groupby('Bairro')
novos_dados_agrupados[['Valor','Area', 'Valor m2']].mean().round(2)
```

```python
novos_dados_agrupados['Valor'].describe()
```

```python
novos_dados_agrupados['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

Podemos observar que tem alguns outliers e mesmo tendo apartamentos de luxos no qual os valores são mais altos que a média, o Gestor informou que nenhum Valor passa de R$ 90.000,00. Logo precisamos remover os outliers.

## Identificando e Removendo Outliers

```python
dados_limpos.loc[dados_limpos['Valor'] > 90000]
```

```python
dados_new = dados_limpos.drop(dados_limpos.loc[dados_limpos['Valor'] > 90000].index)
```
